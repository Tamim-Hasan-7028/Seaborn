# **Regression Plot**


---

## regplot is the AxesSubplot
## FacetGrid is lmplot() or linear model plot.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
tips=pd.read_csv('/content/drive/MyDrive/DataCamp/Data/tips.csv')
tips.head()
```

## **RegPlot**


---



1) Create a basic reg plot

```python
sns.regplot(data=tips,x='total_bill',y='tip',ci=None,color='red'
           ,marker='+',line_kws=dict(color='red',linestyle='--')
           ,scatter_kws=dict(s=100,color='green',alpha=0.5))
plt.show()
```

2) Create a residual plot

```python
sns.residplot(data=tips,x='total_bill',y='tip',color='green'
              ,scatter_kws=dict(s=10,color='green',alpha=0.5))
plt.show()
```

3) Create Polynomial Regression

We have to use order parameter.

```python
#sns.regplot(data=tips,x='total_bill',y='tip',ci=None,order=2)
#plt.show()
```

4) show the regplot without fitted line

```python
sns.regplot(data=tips,x='total_bill',y='tip',ci=None,fit_reg=False)
plt.show()
```

## **lmplot**


---
This is FacetGrid, so we can use row and col option as well


```python
sns.lmplot(data=tips,x='total_bill',y='tip',col='time',row='sex',ci=None,hue='smoker',
           scatter_kws=dict(alpha=0.5))
plt.show()
```
