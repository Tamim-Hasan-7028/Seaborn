```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
mar=pd.read_excel('/content/drive/MyDrive/DataCamp/Data/mart_train_sample.xlsx')
mar.head()
```

## **Scatter Plot**


---
Bivariate & Continuous


1) Create a basic SCATTER PLOT

```python
sns.set_style('whitegrid')
plt.figure(figsize=(15,8))
sns.scatterplot(data=mar,x='Item_MRP',y='Sales',hue='Tier',size='Outlet_Size',
                size_order=['High','Medium','Small'],palette='flare',
                sizes={'High':450,'Medium':250,'Small':150},alpha=0.7)
plt.xlabel('MRP',fontsize=20,fontweight="bold",fontfamily="serif",labelpad=10)
plt.ylabel('Sales',fontsize=20,fontweight="bold",fontfamily="serif",labelpad=10)
plt.title('Scatter Diagram',fontsize=24,fontweight="bold",fontfamily="serif",y=1.02)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(loc="upper right",fontsize=16, bbox_to_anchor=(1.2, 1))
plt.show()
```

2) Grouping basis on a categorical variable using STYLE and Styling Markers as well and Change the Marker size with S arguement

```python
sns.set_style('whitegrid')
plt.figure(figsize=(15,8))
sns.scatterplot(data=mar,x='Item_MRP',y='Sales',style='Outlet_Size'
                ,markers={"High":"^","Small":"v","Medium":"o"},s=150,color='red',alpha=.7)
plt.xlabel('MRP',fontsize=20,fontweight="bold",fontfamily="serif",labelpad=10)
plt.ylabel('Sales',fontsize=20,fontweight="bold",fontfamily="serif",labelpad=10)
plt.title('Scatter Diagram',fontsize=24,fontweight="bold",fontfamily="serif",y=1.02)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(title='Outlet Size', loc="upper right",title_fontsize=18,fontsize=16, 
           bbox_to_anchor=(1.17, 1))
plt.show()
```

## **Line Plot**


---

Bivariate & Continuous. But here one variable is continuous and another one must be Time series variable.

1) Creat a basic line plot and try different estimators

```python
sns.set_style('whitegrid')
plt.figure(figsize=(15,8))
sns.lineplot(data=mar,x="Outlet_Year",y="Sales",ci=None,estimator=None)
plt.xlabel('Year',fontsize=20,fontweight="bold",fontfamily="serif",labelpad=10)
plt.ylabel('Sales',fontsize=20,fontweight="bold",fontfamily="serif",labelpad=10)
plt.title('Line Plot',fontsize=24,fontweight="bold",fontfamily="serif",y=1.02)
plt.xticks(range(1985,2011,2),fontsize=12)
plt.yticks(range(0,5500,500),fontsize=12)
plt.show()
```

2) Grouping using HUE and use different Palettes

```python
sns.set_style('whitegrid')
plt.figure(figsize=(15,8))
sns.lineplot(data=mar,x="Outlet_Year",y="Sales",ci=None
             ,hue='Outlet_Size'
            ,palette=['red','green','blue'])
plt.xlabel('Year',fontsize=20,fontweight="bold",fontfamily="serif",labelpad=10)
plt.ylabel('Sales',fontsize=20,fontweight="bold",fontfamily="serif",labelpad=10)
plt.title('Line Plot',fontsize=24,fontweight="bold",fontfamily="serif",y=1.02)
plt.xticks(range(1985,2011,2),fontsize=12)
plt.yticks(range(0,5500,500),fontsize=12)
plt.legend(title='Outlet Size', loc="upper right",title_fontsize=18,fontsize=16, 
           bbox_to_anchor=(1.17, 1))
plt.show()
```

3) Grouping using Style, use different Marker and Dashes

```python
sns.set_style('whitegrid')
plt.figure(figsize=(15,8))
sns.lineplot(data=mar,x="Outlet_Year",y="Sales",ci=None
             ,style='Outlet_Size',hue='Outlet_Size'
            ,markers=True,dashes=False)
plt.xlabel('Year',fontsize=20,fontweight="bold",fontfamily="serif",labelpad=10)
plt.ylabel('Sales',fontsize=20,fontweight="bold",fontfamily="serif",labelpad=10)
plt.title('Line Plot',fontsize=24,fontweight="bold",fontfamily="serif",y=1.02)
plt.xticks(range(1985,2011,2),fontsize=12)
plt.yticks(range(0,5500,500),fontsize=12)
plt.legend(title='Outlet Size', loc="upper right",title_fontsize=18,fontsize=16, 
           bbox_to_anchor=(1.17, 1))
plt.show()
```

4) Grouping using Size

```python
sns.set_style('whitegrid')
plt.figure(figsize=(15,8))
sns.lineplot(data=mar,x="Outlet_Year",y="Sales",ci=None
            ,size='Outlet_Size',size_order=['High','Medium','Small']
            ,sizes=(.5,5))
plt.show()
```

## **RelPlot**


---



Use col and col wrap

```python
sns.relplot(data=mar,x='Outlet_Year',y='Sales',kind='scatter',col='Outlet_Size',
            hue='Outlet_Size',col_order=['High','Medium','Small'],
            col_wrap=2)
plt.show()
```

Use col and row

```python
sns.relplot(data=mar,x='Outlet_Year',y='Sales',kind='scatter',col='Outlet_Size',
            row='Tier',hue='Outlet_Size',col_order=['High','Medium','Small'])
plt.show()
```
