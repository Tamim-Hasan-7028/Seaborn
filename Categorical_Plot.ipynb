```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
mart=pd.read_excel('/content/drive/MyDrive/DataCamp/Data/supermarket_sales.xlsx')
mart.head()
```

## **Countplot**


---

This is the bar diagram of statistics. Univariate and Categorical plot.

1) Create a basic COUNTPLOT to get number of transaction for each of the product line and change the figure size

```python
sns.set_style('darkgrid')
plt.figure(figsize=(15,10))
sns.countplot(data=mart,x='Product line')
plt.xlabel('Product',fontsize=20,fontweight="bold",fontfamily="serif",labelpad=10)
plt.ylabel('Counts',fontsize=20,fontweight="bold",fontfamily="serif",labelpad=10)
plt.title('Bar Diagram',fontsize=20,fontweight="bold",fontfamily="serif",y=1.01)
plt.xticks(fontsize=12)
plt.yticks(range(0,200,10),fontsize=12)
plt.show()
```

2) Make it horizontal bar plot

```python
sns.set_style('darkgrid')
plt.figure(figsize=(15,8))
g=sns.countplot(data=mart,y='Product line')
g.set(xlabel='Product',ylabel='Counts')
g.set_title('Bar Diagram')
#plt.xticks(rotation=45)
#plt.yticks(rotation=45)
plt.show()
```

3) Add hue to get the count on two categories i.e. Product line and Gender

```python
sns.set_style('whitegrid')
plt.figure(figsize=(15,8))
sns.countplot(data=mart,x='Product line',hue='Gender',hue_order=["Male","Female"],palette='PuBu')
plt.xlabel('Product',fontsize=20,fontweight="bold",fontfamily="serif",labelpad=10)
plt.ylabel('Counts',fontsize=20,fontweight="bold",fontfamily="serif",labelpad=10)
plt.title('Bar Diagram',fontsize=20,fontweight="bold",fontfamily="serif",y=1.01)
plt.xticks(fontsize=12)
plt.yticks(range(0,110,10),fontsize=12)
plt.legend(title="Gender", loc="upper right",title_fontsize=16, fontsize=16, bbox_to_anchor=(1.16, 1))
plt.show()
```

## **Boxplot**


---

Univariate and Continuous plot.

1) Create a basic BOX plot on one NUMERIC variable

```python
sns.set_style('whitegrid')
plt.figure(figsize=(15,8))
sns.boxplot(data=mart,x='Total')
plt.xlabel('Total',fontsize=20,fontweight="bold",fontfamily="serif",labelpad=10)
#plt.ylabel('Count',fontsize=20,fontweight="bold",fontfamily="serif",labelpad=10)
plt.title('Boxplot',fontsize=20,fontweight="bold",fontfamily="serif",y=1.01)
plt.xticks(range(0,1100,50),fontsize=12,rotation=90)
#plt.yticks(range(0,250,10),fontsize=12)
plt.show()
```

2) Create a basic BOX plot on one numeric variable by a CATEGORICAL variable

```python
sns.set_style('whitegrid')
plt.figure(figsize=(15,8))
sns.boxplot(data=mart,x='Payment',y='Total')
plt.xlabel('Payment',fontsize=20,fontweight="bold",fontfamily="serif",labelpad=10)
plt.ylabel('Total',fontsize=20,fontweight="bold",fontfamily="serif",labelpad=10)
plt.title('Boxplot',fontsize=20,fontweight="bold",fontfamily="serif",y=1.01)
plt.xticks(fontsize=16)
plt.yticks(range(0,1100,50),fontsize=12)
plt.show()
```

3) Create a basic BOX plot on one numeric variable by TWO CATEGORICAL variable using HUE attribute with mean

```python
sns.set_style('whitegrid')
plt.figure(figsize=(15,8))
sns.boxplot(data=mart,x='Payment',y='Total',hue='Gender',palette='CMRmap',linewidth=2,
            showmeans=True,meanprops={"marker":"o","markerfacecolor":"white",
                                      "markersize":"5","markeredgecolor":"black"})
plt.xlabel('Payment',fontsize=20,fontweight="bold",fontfamily="serif",labelpad=10)
plt.ylabel('Total',fontsize=20,fontweight="bold",fontfamily="serif",labelpad=10)
plt.title('Boxplot',fontsize=20,fontweight="bold",fontfamily="serif",y=1.01)
plt.xticks(fontsize=16)
plt.yticks(range(0,1100,50),fontsize=12)
plt.legend(title="Gender", loc="upper right",title_fontsize=16, fontsize=16, bbox_to_anchor=(1.16, 1))
plt.show()
```

## **Violin Plot**


---

Univariate and Continuous plot.

1) Create a basic VIOLINPLOT

```python
sns.set_style('whitegrid')
plt.figure(figsize=(15,8))
sns.violinplot(data=mart,x='Total')
plt.xlabel('Total',fontsize=20,fontweight="bold",fontfamily="serif",labelpad=10)
#plt.ylabel('Count',fontsize=20,fontweight="bold",fontfamily="serif",labelpad=10)
plt.title('Boxplot',fontsize=20,fontweight="bold",fontfamily="serif",y=1.01)
plt.xticks(range(0,1100,50),fontsize=12,rotation=90)
#plt.yticks(range(0,250,10),fontsize=12)
plt.show()
```

2) Create a VIOLINPLOT on two categorical and one numeric variable and use split

```python
sns.set_style('whitegrid')
plt.figure(figsize=(15,8))
sns.violinplot(x='Payment',y='Total',hue='Gender',data=mart,split=True,palette='autumn',
               inner='quartile')
plt.xlabel('Payment',fontsize=20,fontweight="bold",fontfamily="serif",labelpad=10)
plt.ylabel('Total',fontsize=20,fontweight="bold",fontfamily="serif",labelpad=10)
plt.title('Violin Plot',fontsize=20,fontweight="bold",fontfamily="serif",y=1.01)
plt.xticks(fontsize=12)
plt.yticks(range(-200,1300,100),fontsize=12)
plt.legend(title="Gender", loc="upper right",title_fontsize=16, fontsize=16, bbox_to_anchor=(1.16, 1))
plt.show()
```

## **CatPlot()**


---

This is a FacetGrid which is a more general function that can create different types of categorical plots based on the kind parameter.

## **Also save the figure**

```python
sns.set_style('whitegrid')
g=sns.catplot(data=mart,x='Payment',y='Total',kind='box',
            hue='Customer type',col='Gender',height=6, aspect=1)  #setting figsize
g.set_xlabels('Payment',fontsize=20,fontweight="bold",fontfamily="serif",labelpad=15)
g.set_ylabels('Total',fontsize=20,fontweight="bold",fontfamily="serif",labelpad=12)
g.fig.suptitle('Boxplot',fontsize=20,fontweight="bold",fontfamily="serif",y=1.05)
g.set_titles('{col_name}',size=16)
g.set_xticklabels(fontsize=16)
plt.yticks(range(0,1100,50),fontsize=12)
plt.savefig("Boxplot1.jpg", dpi=300)  #dpi is for resulation
plt.show()
```

```python
sns.set_style('whitegrid')
g=sns.catplot(data=mart,x='Payment',y='Total',kind='box',
            row='Customer type',col='Gender',height=5, aspect=1)  #setting figsize
g.set_xlabels('Payment',fontsize=20,fontweight="bold",fontfamily="serif",labelpad=15)
g.set_ylabels('Total',fontsize=20,fontweight="bold",fontfamily="serif",labelpad=12)
g.fig.suptitle('Boxplot',fontsize=20,fontweight="bold",fontfamily="serif",y=1.03)
g.set_titles('{col_name} {row_name}',size=16)
g.set_xticklabels(fontsize=16)
plt.yticks(range(0,1100,50),fontsize=12)
plt.savefig("Boxplot.png", dpi=300, bbox_inches="tight") #dpi is for resulation
plt.show()
```
