## **Hist Plot**


---

Is the Histogram. Univariate & Continuous.

1) Create a basic HISTOGRAM on X or Y axis

binwidth define the range of per group (bar).

bins define the how many groups (bars) will be there.

Total Range/Bins=Binwidth

```python
sns.set_style('whitegrid')
plt.figure(figsize=(15,8))
sns.histplot(data=mart,x='Total',binwidth=100,kde=True)
plt.xlabel('Total',fontsize=20,fontweight="bold",fontfamily="serif",labelpad=10)
plt.ylabel('Count',fontsize=20,fontweight="bold",fontfamily="serif",labelpad=10)
plt.title('Histogram',fontsize=20,fontweight="bold",fontfamily="serif",y=1.01)
plt.xticks(range(0,1200,100),fontsize=12)
plt.yticks(range(0,250,10),fontsize=12)
plt.show()
```

2) create 5 bars


```python
plt.figure(figsize=(15,8))
sns.histplot(data=mart,x='Total',bins=5)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

3) Use a categorical variable in HUE and STACK it using MULTIPLE agruement

```python
sns.set_style('whitegrid')
plt.figure(figsize=(15,8))
sns.histplot(data=mart,x='Total',binwidth=100,hue='Payment',multiple='stack',kde=True)
plt.xlabel('Total',fontsize=20,fontweight="bold",fontfamily="serif",labelpad=10)
plt.ylabel('Count',fontsize=20,fontweight="bold",fontfamily="serif",labelpad=10)
plt.title('Histogram',fontsize=20,fontweight="bold",fontfamily="serif",y=1.01)
plt.xticks(range(0,1200,100),fontsize=12)
plt.yticks(range(0,250,10),fontsize=12)
legend_labels=['Credit Card', 'E-Wallet','Cash']
plt.legend(title="Payment",labels=legend_labels, loc="upper right",title_fontsize=18,
           fontsize=16, bbox_to_anchor=(1.2, 1))
plt.show()
```

## **Kernal Density Estimation Plot**


---

Univariate & Continuous.

1) Create a basic KDE plot for Total column

```python
sns.kdeplot(data=mart, x='Total')
plt.show()
```

2) Group the KDE on a category variable

```python
sns.kdeplot(data=mart,x='Total',hue='Payment',multiple='stack')
plt.show()
```

3) Use log scaling to map the variable in KDE

```python
sns.kdeplot(data=mart,x='Total',log_scale=True)
plt.show()
```

4) Change styling of hued KDE using linewidth, palette, alpha

```python
sns.kdeplot(data=mart,x='Total',hue='Payment',multiple='stack',linewidth=1,palette='Dark2',alpha=0.5)
plt.show()
```

## **Emparical Cumalative Density Function Plot**


---



1) Create ECDF plot with gross income and gender and find the total count.

```python
sns.ecdfplot(data=mart,y='gross income',hue='Gender',stat='count')
plt.show()
```

## **Displot**


---

By defult it gives histplot

1) Create displot with total where kde and rug is available and payment, branch, gender in hue, col, row respectively

```python
sns.set_style('whitegrid')
g=sns.displot(data=mart,x='Total',kde=True,rug=True,rug_kws={'height':0.05},
            hue='Payment',multiple='stack',col='Branch',row='Gender',bins=10,
            height=7,aspect=1,col_order=['A','B','C'])
g.set_xlabels('Total',fontsize=20,fontweight="bold",fontfamily="serif",labelpad=15)
g.set_ylabels('Count',fontsize=20,fontweight="bold",fontfamily="serif",labelpad=12)
g.fig.suptitle('Histogram',fontsize=20,fontweight="bold",fontfamily="serif",y=1.03)
g.set_titles('{col_name} {row_name}',size=16)
g.set_xticklabels(fontsize=16)
plt.yticks(range(0,50,5),fontsize=12)
plt.show()
```

2) Create displot with total where ecdf is on and payment, branch, gender in hue, col, row respectively

```python
sns.set_style('whitegrid')
sns.displot(data=mart,y='Total',kind='ecdf',hue='Payment',col='Branch',row='Gender')
g.set_xlabels('Proportion',fontsize=20,fontweight="bold",fontfamily="serif",labelpad=15)
g.set_ylabels('Total',fontsize=20,fontweight="bold",fontfamily="serif",labelpad=12)
g.fig.suptitle('Cumalative',fontsize=20,fontweight="bold",fontfamily="serif",y=1.03)
g.set_titles('{col_name} {row_name}',size=16)
g.set_xticklabels(fontsize=16)
plt.yticks(fontsize=12)
plt.show()
```
